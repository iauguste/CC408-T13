#### Ejercicio 1
Usando la API de Mercado Libre, obtener los ítems de una consulta de búsqueda. Pueden buscar cualquier producto de su interés.

```python
# Resolver acá

import requests

# Así se ve la búsqueda en el navegador:
# https://listado.mercadolibre.com.ar/cuadros-decorativos#D[A:cuadros%20decorativos]

buscar = "cuadros decorativos"
url = "https://api.mercadolibre.com/sites/{}/search?q={}".format("MLA", buscar.replace(" ", "%20"))

# Hacemos el pedido o request y obtenemos la response
response = requests.request("GET", url)
print(response) 

# Vemos el texto
response.text

```

#### Ejercicio 2
Guarden los precios de los ítems obtenidos en un dataframe y calculen el precio promedio, el mínimo y el máximo. 

```python
# Resolver acá

# guardamos en un diccionario (json)
data_cuadros = response.json()

import pandas as pd  

# guardo titulo y precio en una lista
productos = []
for item in data_cuadros['results']:
    productos.append({'Producto': item['title'], 'Precio': item['price']})

# Creo dataframe
df_precios = pd.DataFrame(productos)

print(df_precios)
```

```python
#guardo los valores en variables
mean_precio = df_precios['Precio'].mean()
max_precio = df_precios['Precio'].max()
min_precio = df_precios['Precio'].min()

#imprimo los resultados redondeados
print("Media de precios: ", round(mean_precio))
print("Precio máximo: ", round(max_precio))
print("Precio mínimo: ", round(min_precio))
```

#### Ejercicio 3
Armen un histograma de los precios. ¿Ven algún <i>outlier<i>? 
Nota: pueden usar la librería de Matplotlib o la de Pandas.

```python
# Resolver acá

import matplotlib.pyplot as plt

# histograma de precios
df_precios['Precio'].hist(bins=30, edgecolor='black')

# etiquetas
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
plt.title('Histograma de Precios')

# Mostrar el gráfico
plt.show()
```

Podemos ver que los precios se concentran entre los $10000 y los $24000, después están mas dispersos. Podríamos decir que los cuadros a partir de los $45000 son outliers.

Realicen alguna consulta adicional utilizando la API de Mercado Libre (debe ser alguna consulta que no se haya visto en clase. Por ejemplo, obtener los ítems de un vendedor en particular, obtener los productos de una categoría u otros).
```python
# Resolver acá

#vamos a buscar las carpas más vendidas y calcular el descuento promedio que tienen esos productos 

# Así se ve la búsqueda en el navegador:
# https://listado.mercadolibre.com.ar/carpas

buscar = "carpas"
url = "https://api.mercadolibre.com/sites/{}/search?q={}".format("MLA", buscar.replace(" ", "%20"))

# Hacemos el pedido o request y obtenemos la response
response = requests.request("GET", url)
print(response)

# Vemos el texto
response.text

# guardamos en un diccionario (json)
data_carpas = response.json()

# guardo titulo y precio en una lista
carpas = []
for item in data_carpas['results']:
    carpas.append({'Producto': item['title'], 'Precio final': item['price'], "Precio original" : item["original_price"]})

# Creo dataframe
df_carpas = pd.DataFrame(carpas)

print(df_carpas)

# Como vimos que las que no tienen descuento tienen "Precio original" como NA,
# los reemplazamos con los valores de "Precio final"
df_carpas['Precio original'].fillna(df_carpas['Precio final'], inplace=True)

# Queremos ver que porocentaje de descuento tiene cada producto, lo guardamos en una nueva columna
df_carpas['Porcentaje de Descuento'] = ((df_carpas['Precio original'] - df_carpas['Precio final']) / df_carpas['Precio original']) * 100

print(df_carpas)

#queremos ver el promedio de descuento en los primeros 50 listings de carpas
promedio_descuento = df_carpas['Porcentaje de Descuento'].mean()

print(promedio_descuento)
#El descuento promedio es de aproximadamente 15%
```

De la página de noticias del [diario La Nación](https://www.lanacion.com.ar/), utilicen herramientas de web scraping para obtener los links de las noticias de la portada. Guarden los links obtenidos en un dataframe y expórtenlo a un archivo de excel.

Nota 1: es posible que logren obtener los links a las noticias sin el dominio: "https://www.lanacion.com.ar/". De ser así, concatenen el dominio a la ruta del link obtenido, tal que se obtenga un link al que se pueda acceder. Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
# Resolver acá
import requests #html requestor
from bs4 import BeautifulSoup #html parser

url = "https://www.lanacion.com.ar/"

r = requests.get(url) #traigo el contenido del html
contenido = r.content

soup = BeautifulSoup(contenido, "html.parser")
soup
```

```python
 # Buscamos las etiquetas <section> que tengan la clase "description-container"
sections = soup.find_all('section', class_='description-container')
    
# Extraemos el valor del atributo 'href' de cada sección
section_links = []
for section in sections:
        href = section.get('href')
        if href:
            full_link = 'https://www.lanacion.com.ar' + href
            section_links.append(full_link)
    
# Imprimimos los links obtenidos
for link in section_links:
        print(link)
```
